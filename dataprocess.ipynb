```python

import xgboost as xgb
from sklearn.model_selection import cross_val_score
xgbmodel = xgb.XGBRegressor(max_depth=5, n_estimators = 400)
xgbmodel.fit(train_dummies, y)
np.sqrt(-cross_val_score(xgbmodel, train_dummies, y, cv=5, scoring = "neg_mean_squared_error")).mean()
```

```python
train_predict = xgbmodel.predict(train_dummies)
test_predict = xgbmodel.predict(test_dummies)
```

```python
q1 = pd.DataFrame(test_predict).quantile(0.0042)
pre_df = pd.DataFrame(test_predict)
pre_df["SalePrice"] = test_predict
pre_df = pre_df[["SalePrice"]]
pre_df.loc[pre_df.SalePrice <= q1[0], "SalePrice"] = pre_df.loc[pre_df.SalePrice <= q1[0], "SalePrice"] *0.96
test_predict = np.array(pre_df.SalePrice)
sample_submission["SalePrice"] = np.exp(test_predict)-1
sample_submission.to_csv("2.csv", index=False)
```
